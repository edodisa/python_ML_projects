# Breast Cancer Prediction with NN

In this project, using the **Winsconsis Breast Cancer Dataset** we will use Keras to implement an Artificial Neural Network to predict breast cancer.

```python
import pandas as pd
```

### Load the dataset

```python
data = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/wdbc.data",
                           names=["id","diagnosis","radius_mean","texture_mean","perimeter_mean","area_mean","smoothness_mean","compactness_mean","concavity_mean","concave points_mean","symmetry_mean","fractal_dimension_mean","radius_se","texture_se","perimeter_se","area_se","smoothness_se","compactness_se","concavity_se","concave points_se","symmetry_se","fractal_dimension_se","radius_worst","texture_worst","perimeter_worst","area_worst","smoothness_worst","compactness_worst","concavity_worst","concave points_worst","symmetry_worst","fractal_dimension_worst"])

data.head(5)
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X = data.drop(['diagnosis','id'],axis=1).values
y = data['diagnosis'].values
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3)
```

## Label encoding

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y_train = le.fit_transform(y_train)
y_test = le.transform(y_test)
```

## Feature scaling

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
X_train = ss.fit_transform(X_train)
X_test = ss.transform(X_test)
```

```python
X_train.shape
```

# NN with one hidden layer

We can now build our ANN. Our first model will have:

* 30 nodes of input, equal to the number of feature of the dataset;
* 12 hidden nodes;
* 1 output node, since it is a classification problem.

We use the **RELU** activation function for the hidden layer and for the output layer we use a **sigmoid**, since it is a classification problem.

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.initializers import RandomUniform
```

```python
model = Sequential()
model.add(Dense(12, input_dim=X_train.shape[1], activation="relu"))
model.add(Dense(1, activation="sigmoid"))
```

By training the NN, we specify the **Stochastic Gradient Descent** as the iterative method for optimizing the objective function, and the **Binary Cross Entropy** as loss function.

```python
model.compile(optimizer='sgd', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.summary()
```

## Train the model

```python
model.fit(X_train, y_train, epochs=100)
```

## Check the accuracy 

```python
loss, acc = model.evaluate(X_test, y_test)
print("Loss on test set: %.4f" % loss)
print("Accuracy on test set: %.4f" % acc)
```

# Deep NN

We build now a Deep Neural Network, a NN with more than a hidden layer.
Our model will have:

* 30 nodes of input, equal to the number of feature of the dataset;
* 12 nodes on the first hidden layer;
* 4 nodes on the second hidden layer;
* 1 output node, since it is a classification problem.

We use the **RELU** activation function for the hidden layers and for the output layer we use a **sigmoid**, since it is a classification problem.

```python
model = Sequential()
model.add(Dense(12, input_dim=X_train.shape[1], activation="relu"))
model.add(Dense(4, activation="relu"))
model.add(Dense(1, activation="sigmoid"))

model.compile(optimizer='sgd', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(X_train, y_train, epochs=100)
```

## Check the accuracy

```python
loss, acc = model.evaluate(X_test, y_test)
print("Loss on test set: %.4f" % loss)
print("Accuracy on test set: %.4f" % acc)
```

# Leaky RELU

We now implement the same model but using the Leaky RELU as activation function for the hidden layers.

```python
from keras.layers import LeakyReLU

model = Sequential()

model.add(Dense(12, input_dim=X_train.shape[1]))
model.add(LeakyReLU(alpha=0.01))
model.add(Dense(4))
model.add(LeakyReLU(alpha=0.01))
model.add(Dense(1, activation='sigmoid'))

model.compile(optimizer='sgd', loss='binary_crossentropy', metrics=['accuracy'])
model.fit(X_train, y_train, epochs=100)
```

## Check the accuracy

```python
loss, acc = model.evaluate(X_test, y_test)
print("Loss on test set: %.4f" % loss)
print("Accuracy on test set: %.4f" % acc)
```

# Confusion Matrix

In breast cancer prediction, we should consider that misdiagnosing a malignant tumor as benign is much more serious than classifying a benign tumor as malignant, a problem common in a lot of classification problems.
In these cases we can use the classification matrix which allows us to understand where our model has made mistakes.

```python
from sklearn.metrics import confusion_matrix

y_pred = (model.predict(X_test) > 0.5).astype("int32")
cm = confusion_matrix(y_test, y_pred)
cm
```
